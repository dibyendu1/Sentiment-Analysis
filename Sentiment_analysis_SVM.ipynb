```python
from nltk.corpus import stopwords
from nltk.tokenize import sent_tokenize
import nltk
import string
from nltk import word_tokenize
from nltk.corpus import stopwords
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
```

```python
# data = np.genfromtxt('IMDB Dataset.csv', delimiter=',')
data = pd.read_csv('IMDB Dataset.csv')
size = 2000
```

```python
for i in range(size):
    row = data['review'].iloc[i]
    print(i)
    sentences = sent_tokenize(row)
    sent_row = []
    word_sent = []
    stop_words = stopwords.words()
#     print(sentences)
    for sentence in sentences:
#         print(sentence)
        word_sent = word_tokenize(sentence)
        no_punct_words = []
        for word in word_sent:
#             print(word)
            if word not in stop_words:
                no_punct_words.append(word)
#         print(no_punct_words)
        no_punct_text = " ".join([char for char in no_punct_words if char not in string.punctuation])
#         print(no_punct_text)
        sent_row.append(no_punct_text)
    row = '.'.join(sent_row)
    data['review'].iloc[i] = row
```

```python
labels = []
for i in range(size):
    if data['sentiment'].iloc[i] == 'positive':
        labels.append(1)
    else:
        labels.append(0)
labels = np.array(labels)
print(labels.shape)
```

```python
cv = CountVectorizer()
full_list = []
for i in range(size):
    full_list.append(data['review'].iloc[i])
# print(full_list)
```

```python
BoW = cv.fit_transform(full_list).toarray()
print(type(BoW))
print(cv.vocabulary_)
```

```python
column_names = cv.get_feature_names()
df = np.zeros((size, len(column_names)))
for row in range(len(BoW)):
    for col in column_names:
        df[row,:] += BoW[row]*cv.vocabulary_[col]
```

```python
df_features = pd.DataFrame(df, columns=column_names)
df_features.info()
```

```python
df_target = np.ravel(labels)
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_features, df_target, test_size=0.30, random_state=101)
```

```python
model = SVC()
```

```python
model.fit(X_train,y_train)
```

```python
predictions = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_test,predictions))
```

```python
print(classification_report(y_test,predictions))
```
